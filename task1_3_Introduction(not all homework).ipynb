```python
from random import randint
lis = []
for i in range(100):
    lis.append(randint(0, 100))
print(lis)
print(sum(lis))
print(max(lis))
pr = 1
for k in lis:
    pr = pr * k
print(pr)
print(lis.index(max(lis)))
print(lis.index(min(lis)))
i = lis.index(max(lis))
m = (lis.index(min(lis)))
lis[i], lis[m] = lis[m], lis[i]
print(lis)
print(set(lis))
```
